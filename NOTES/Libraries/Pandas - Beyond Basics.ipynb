```python
### import pandas as pd 
data = [
    ['110', 'g', 201751, 0],
    ['110', 'g', 201752, 0],
    ['110', 'g', 201801, 1],
    ['110', 'a', 201802, 2],
    ['110', 'a', 201803, 3],
    ['110', 'a', 201806, 2],
    ['110', 'a', 201807, 3],
    ['110', 'a', 201808, 10],
    ['110', 'a', 201809, 2],
    ['110', 'a', 201810, 10],
    ['110', 'a', 201811, 1],
    ['113', 'g', 201930, 4],    
    ['111', 'g', 201902, 3],
    ['111', 'g', 201903, 360],
    ['111', 'g', 201905, 3],
    ['111', 'g', 201906, 4],
    ['111', 'g', 201901, 2],
    ['112', 'c', 201905, 60],
    ['114', 'c', 201905, 90],
    ['114', 'a', 201801, 0],
    ['114', 'd', 201948, 0],
]

pd_df = pd.DataFrame(data, columns = ['Material_ID', 'PL3_Category', 'Week_ID', "Sales"])
pd_df
```

```python
product_cols = 'Material_ID'
time_var = 'Week_ID'
target_var = 'Sales'
time_threshold = 3
target_threshold = 5

agg_df = pd_df.groupby([product_cols] + [time_var])[target_var].sum().reset_index()
agg_df.sort_values([product_cols] + [time_var], ascending=[True, True])
agg_df
```

```python
agg_df = agg_df[agg_df[target_var] <= target_threshold]
agg_df
```

```python
# %48 when week transition from xxxx52 to xxxx01.
agg_df['week_difference'] = np.where((agg_df[product_cols] == agg_df[product_cols].shift(-1)),\
                                    ((agg_df[time_var].shift(-1) - agg_df[time_var]) % 48),\
                                    0)
agg_df
```

```python
agg_df['cons_week_flag'] = np.where(agg_df['week_difference'] == 1, 1, 0)
agg_df
```

```python
agg_df['max_cons_week'] = agg_df['cons_week_flag'] * (agg_df.groupby((agg_df['cons_week_flag'] != agg_df['cons_week_flag'].shift()).cumsum()).cumcount() + time_threshold - 1)
agg_df
```

```python
velocityFlagDf = agg_df.groupby(product_cols)['max_cons_week'].max().reset_index()
velocityFlagDf['lowVelocityFlag'] = np.where(velocityFlagDf['max_cons_week'] >= time_threshold, 1, 0)
velocityFlagDf
```

```python
velocityFlagDf.drop(['max_cons_week'], axis=1, inplace=True)
velocityFlagDf
```

```python
pd_df = pd_df.merge(velocityFlagDf, how='left', left_on=product_cols, right_on=product_cols)
pd_df
```
